**¿Qué es la Tuberculosis?**  

Es una enfermedad grave que afecta principalmente los pulmones. Los microbios que causan la tuberculosis son un tipo de bacteria.  

La tuberculosis puede contagiarse cuando una persona con la enfermedad tose, estornuda o canta. Esto puede hacer que se liberen gotitas diminutas con microbios en el aire. Después, otra persona puede respirar las gotitas, y los microbios pueden ingresar en sus pulmones.  

La tuberculosis se trasmite fácilmente cuando las personas se reúnen en multitudes o si viven con muchas otras personas. Las personas con VIH/SIDA y otras personas que tienen un sistema inmunitario debilitado corren un mayor riesgo de contagiarse con tuberculosis que aquellas que tienen un sistema inmunitario normal.

**Riesgo de infección**

* Tienen VIH/SIDA
* Usan medicamentos intravenosos
* Están en contacto con personas infectadas
* Son de un país donde la tuberculosis es común, como varios países de América Latina, África y Asia
* Viven o trabajan en zonas donde la tuberculosis es común, como las cárceles o los asilos de ancianos y convalecientes
* Trabajan en la atención médica y tratan personas con alto riesgo de tuberculosis
* Tienen niños que están expuestos a adultos con riesgo de tuberculosis

## Modificar la estructura de los datos

```python
who.head()
```

```python
population.head()
```

```python
who2 = who.melt(id_vars=["country","year","iso2","iso3"])
who2.head(10)
```

```python
# Añadir la columna que indique el género
who2["gender"] = np.where(who2["variable"].str.contains("m"),"masculino","femenino")
who2.head(10) #masculino
#who2.sort_values(by="gender").head(250) #femenino
```

```python
# Función para asignar el grupo de edad
def asignar_groupedad(valor):
  if "014" in valor:
    return "0-14"
  elif "1524" in valor:
    return "15-24"
  elif "2534" in valor:
    return "25-34"
  elif "3544" in valor:
    return "35-44"
  elif "4554" in valor:
    return "45-54"
  elif "5564" in valor:
    return "55-64"
  elif "65" in valor:
    return "65+"

```

```python
who2["agegroup"] =who2["variable"].apply(asignar_groupedad)
who2.head()
```

```python
#Función para asignar los métodos.
#the method of diagnosis (rel = relapse,
#sn = negative pulmonary smear,
#sp = positive pulmonary smear, ep = extrapulmonary),

def asignar_methodo(valor):
  if "rel" in valor:
    return "recaida"
  if "sn" in valor:
    return "esputo pulmonar negativo"
  if "sp" in valor:
    return "esputo pulmonar positivo"
  if "ep" in valor:
    return "extrapulmonar"
```

```python
who2["metod"] =who2["variable"].apply(asignar_methodo)
who2.head()
```

```python
who2 = who2.drop(columns=["variable"])
```

```python
who2.head()
```

```python
who2 = who2.rename(columns={"value":"cases"})
```

```python
who2.head()
```

```python
# Generamos un join con population
df = pd.merge(left= who2,right= population, how ="inner", on =["country","year"])
df.head(100)
```

```python
#Afghanistan 2000 masculino 0-14 pulmonar positivo -1000 cases -100_000 poblacion
#Afghanistan 2000 femenino 0-14 pulmonar positivo -500 cases -100_000 poblacion

#Dataframe agrupe por país y año.
#Dataframe by_country_year
#Afghanistan año 2000 cases 1500 poblacion 100_000 => i=100_000 *cases/poblacion
#Afghanistan año 2001 cases 1000 poblacion 120_000  => i=100_000 *cases/poblacion
```

Calcula el porcentaje de cambio en la incidencia entre 1995 (año base ONU) y el año más reciente.



Reducción en la tasa de incidencia del 50% 1995 vs 2013.

```python
# Filtrar 1995 y 2013
base = by_country_year[by_country_year['year'] == 1995]
latest = by_country_year[by_country_year['year'] == by_country_year['year'].max()]

# Merge para comparación
evolucion = pd.merge(base, latest, on="country", suffixes=("_1995", "_latest"))

# Calcular cambio porcentual
evolucion["cambio_pct"] = 100 * (evolucion["incidencia_1995"] - evolucion["incidencia_latest"]) / evolucion["incidencia_1995"]

# Marcar si cumple la meta de reducción del 50%
evolucion["cumple_objetivo"] = evolucion["cambio_pct"] >= 50

# Mostrar top 10 mejoras
evolucion.sort_values("cambio_pct", ascending=False).head(10)
```

Incidencia global anual



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Agrupar incidencia por año (suma de casos y población)
incidencia_anual = df.groupby("year").agg(
    casos_totales=("cases", "sum"),
    poblacion_total=("population", "sum")
).reset_index()

incidencia_anual["incidencia"] = 100000 * incidencia_anual["casos_totales"] / incidencia_anual["poblacion_total"]

# Gráfico de línea para incidencia a lo largo de los años
plt.figure(figsize=(10,6))
sns.lineplot(data=incidencia_anual, x="year", y="incidencia", marker="o")
plt.title("Incidencia global de tuberculosis (1995-2013)")
plt.ylabel("Casos por 100,000 habitantes")
plt.xlabel("Año")
plt.grid(True)
plt.show()
```

Análisis por género.

```python
# Agrupar por año y género
incidencia_genero = df.groupby(["year", "gender"]).agg(
    casos=("cases", "sum"),
    poblacion=("population", "sum")
).reset_index()

incidencia_genero["incidencia"] = 100000 * incidencia_genero["casos"] / incidencia_genero["poblacion"]

# Gráfico líneas separadas por género
plt.figure(figsize=(10,6))
sns.lineplot(data=incidencia_genero, x="year", y="incidencia", hue="gender", palette={"masculino": "blue", "femenino": "pink"})
plt.title("Incidencia de tuberculosis por género (1995-2013)")
plt.ylabel("Casos por 100,000 habitantes")
plt.xlabel("Año")
plt.grid(True)
plt.show()
```

Análisis por grupo de Edad.

```python
# Sumar casos y población por grupo de edad total 1995-2013
edad_total = df.groupby("agegroup").agg(
    casos=("cases", "sum"),
    poblacion=("population", "sum")
).reset_index()

edad_total["incidencia"] = 100000 * edad_total["casos"] / edad_total["poblacion"]

# Barras de incidencia por grupo de edad
plt.figure(figsize=(10,6))
sns.barplot(data=edad_total.sort_values("incidencia", ascending=False),
            x="agegroup", y="incidencia", hue="agegroup", palette="mako", legend=False)
plt.title("Incidencia de tuberculosis por grupo de edad (1995-2013)")
plt.xlabel("Grupo de edad")
plt.ylabel("Casos por 100,000 habitantes")
plt.show()
```

Análisis por método de diagnóstico

```python
# Casos por método (total acumulado)
metodo_total = df.groupby("metod").agg(
    casos=("cases", "sum")
).reset_index()

# Barras
plt.figure(figsize=(8,5))
sns.barplot(data=metodo_total.sort_values("casos", ascending=False),
            x="metod", y="casos", hue="metod", palette="Set2", legend=False)
plt.title("Casos de tuberculosis por método diagnóstico (1995-2013)")
plt.xticks(rotation=30)
plt.ylabel("Número de casos")
plt.xlabel("Método diagnóstico")
plt.show()
```
